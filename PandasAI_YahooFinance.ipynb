```python
from pandasai.connectors.yahoo_finance import YahooFinanceConnector
from pandasai import SmartDataframe

yahoo_connector = YahooFinanceConnector("BTC-USD")
df = SmartDataframe(yahoo_connector)
#prices depicted should be multipied by 1000 for BTC.
response = df.chat("Plot the chart of the stock")
print(response)
```

```python
from pandasai import Agent

yahoo_connector = YahooFinanceConnector("BTC-USD")

agent = Agent(yahoo_connector)
```

```python
response = agent.chat("Plot the chart of the stock close price")
```

```python
result = agent.chat("Can you also add the 20 and 200 day moving average at 1000 days?")
```
