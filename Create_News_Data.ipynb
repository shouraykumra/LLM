```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, "html.parser")

list_of_article_links = []

links = soup.find_all("a", class_='mega-item-header-link')
# Extract the URL of the first link
for link in range(len(links)):
  article_link = links[link]['href']
  list_of_article_links.append(article_link)

print(list_of_article_links)
```
